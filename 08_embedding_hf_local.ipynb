```python
from langchain_huggingface import HuggingFaceEmbeddings

# Initialize the HuggingFaceEmbeddings model.
# This component allows you to load and use embedding models from the Hugging Face Hub.
# These models are designed to convert text into numerical vectors (embeddings)
# that capture the semantic meaning of the text.
# - 'model_name': Specifies the exact pre-trained model to load from the
#                 Hugging Face 'sentence-transformers' collection.
#                 'sentence-transformers/all-MiniLM-L6-v2' is a popular and efficient
#                 model for generating sentence embeddings.
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

# Define a list of documents (strings) that we want to embed.
# Each string represents a distinct piece of text.
documents = [
    "Tokyo is the capital of Japan",
    "Rome is the capital of Italy",
    "Berlin is the capital of Germany"
]

# Generate embeddings for the list of documents.
# The 'embed_documents' method takes a list of strings and processes them
# through the loaded Hugging Face embedding model. It returns a list of
# numerical vectors, where each vector corresponds to the semantic representation
# of the respective document in the input list.
vector = embedding.embed_documents(documents)

# Print the resulting list of embedding vectors.
# The 'vector' variable will contain a list of lists of floats,
# where each inner list is the embedding vector for one of the documents.
print(str(vector))
```
